# Randomly initiated 2-way k-means clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def initialize_centroids(data, k):
    # Randomly initialize k centroids
    centroids_indices = np.random.choice(len(data), k, replace=False)
    centroids = data.iloc[centroids_indices]
    return centroids


def assign_to_clusters(data, centroids):
    # Assign each data point to the closest centroid
    if isinstance(data, pd.DataFrame):
        data_values = data.values
    else:
        data_values = data

    if isinstance(centroids, pd.DataFrame):
        centroids_values = centroids.values
    else:
        centroids_values = centroids

    distances = np.linalg.norm(data_values[:, np.newaxis, :] - centroids_values, axis=2)
    clusters = np.argmin(distances, axis=1)
    return clusters


def update_centroids(data, clusters, k):
    # Update centroids based on the mean of data points in each cluster
    centroids = np.array([data[clusters == i].mean(axis=0) for i in range(k)])
    return centroids


def compute_between_cluster_distance(data, clusters, centroids):
    # Compute the between-cluster distance objective (JD)
    n = len(data)
    JD = 0.0
    for i in range(len(centroids)):
        for j in range(i + 1, len(centroids)):
            d_ij = np.sum((centroids[i] - centroids[j]) ** 2)
            JD += (len(data[clusters == i]) * len(data[clusters == j]) / n) * (2 * d_ij / (len(data[clusters == i]) * len(data[clusters == j])) - d_ij / len(data[clusters == i]) ** 2 - d_ij / len(data[clusters == j]) ** 2)

    return JD


def two_way_kmeans(data, k=2, num_iterations=100, stability_threshold=1):
    # Randomly initialize centroids
    centroids = initialize_centroids(data, k)
    prev_clusters = np.zeros(len(data))

    for _ in range(num_iterations):
        # Assign data points to clusters based on current centroids
        clusters = assign_to_clusters(data, centroids)

        # Update centroids
        centroids = update_centroids(data, clusters, k)

        # Check for stability and break the loop if stable
        if np.array_equal(clusters, prev_clusters):
            break
        prev_clusters = clusters.copy()

    # Compute the between-cluster distance objective (JD)
    JD = compute_between_cluster_distance(data, clusters, centroids)

    return clusters, centroids, JD
    
    
# Example usage:
# Load the CSV file
file_path = 'diabetes_NHANES_1999_2004.csv'
data = pd.read_csv(file_path)

x = 'RIDAGEYR'
y = 'BMXWAIST'

# x = 'BMXHT'
# y = 'BMXWT'

# Select two variables for 2-way k-means clustering
selected_data = data[[x, y]]

# Remove the rows with NA data 
remove_na_selected_data = selected_data.dropna(axis=0)

# Perform 2-way k-means clustering
clusters, centroids, JD = two_way_kmeans(remove_na_selected_data)

# Visualize the results (scatter plot)
plt.scatter(remove_na_selected_data[x], remove_na_selected_data[y], c=clusters, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroids')
plt.xlabel('Age')
plt.ylabel('Waist')
plt.title('2-way K-means Clustering')
plt.legend()
plt.show()
```

# PCA-guided 2-way k-means clustering

```python
def mean_center_data(data):
    #calculate mean of each column
    M = np.mean(data, axis=0)

    #calculate centered data matrix 
    centered_data = data - M 

    return centered_data


def pca(centered_data, n_components):
    #calculate covariance matrix
    cov_matrix = np.cov(centered_data, rowvar=False)

    #compute eigenvalues and eigenvectors of matrix
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

    #find the indices for the eigenvalues sorted in descending order
    sorted_idx = np.argsort(eigenvalues)[::-1]
    
    #sort the eigenvalues with the indices
    sorted_eigenvalues = eigenvalues[sorted_idx]

    #sort the eigenvalues in descending order
    sorted_eigenvectors = eigenvectors[:,sorted_idx]

    #top n eigenvectors
    top_n_eigenvectors = sorted_eigenvectors[:, :n_components]
    
    return top_n_eigenvectors


def PCA_guided_kmeans(data, num_iterations=100, stability_threshold=1):
    # Center the data
    centered_data = mean_center_data(data)

    # Perform PCA to get the first principal component
    top_1_eigenvector = pca(centered_data, 1)

    # Assign data points to clusters based on the sign of the first principal component
    clusters = np.sign(np.dot(centered_data, top_1_eigenvector))

    # Define indices for the two clusters
    cluster_1_indices = np.where(clusters < 0)[0]
    cluster_2_indices = np.where(clusters >= 0)[0]

    # Initialize centroids as the mean of each cluster
    centroid_1 = np.mean(data.iloc[cluster_1_indices], axis=0).values
    centroid_2 = np.mean(data.iloc[cluster_2_indices], axis=0).values

    centroids = np.vstack((centroid_1, centroid_2))
    prev_clusters = np.zeros(len(data))

    for _ in range(num_iterations):
        # Assign data points to clusters based on current centroids
        clusters = assign_to_clusters(data, centroids)

        # Update centroids based on the original data points
        centroids = update_centroids(data, clusters, 2)

        # Check for stability and break the loop if stable
        if np.array_equal(clusters, prev_clusters):
            break
        prev_clusters = clusters.copy()

    # Compute the between-cluster distance objective (JD)
    JD = compute_between_cluster_distance(data, clusters, centroids)

    return clusters, centroids, JD


# Remove the rows with NA data 
remove_na_selected_data = selected_data.dropna(axis=0)

# Perform PCA-guided 2-way k-means clustering
clusters, centroids, JD = PCA_guided_kmeans(remove_na_selected_data)

# Visualize the results (scatter plot)
plt.scatter(remove_na_selected_data[x], remove_na_selected_data[y], c=clusters, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroids')
plt.xlabel('Age')
plt.ylabel('Waist')
plt.title('PCA-guided 2-way K-means Clustering')
plt.legend()
plt.show()
```

# Visualization to show how PCA and K-means are associated

```python
# Perform PCA-guided 2-way k-means clustering
clusters, centroids, JD = two_way_kmeans(remove_na_selected_data)

# Get the first PCA component values
pca_values = np.dot(mean_center_data(remove_na_selected_data), pca(remove_na_selected_data, 1))

# Create a scatter plot
plt.scatter(remove_na_selected_data.index, pca_values, c=clusters, cmap='viridis')
plt.xlabel('Index')
plt.ylabel('First PCA Component Value')
plt.title('Randomly initiated 2-way K-means Clustering')
plt.show()
```

```python
import timeit

# Function to measure the execution time of PCA_guided_kmeans
def time_pca_guided_kmeans():
    PCA_guided_kmeans(remove_na_selected_data)

# Function to measure the execution time of two_way_kmeans
def time_two_way_kmeans():
    two_way_kmeans(remove_na_selected_data)

# Number of experiments
num_experiments = 100

# Measure the time for PCA_guided_kmeans
pca_guided_kmeans_time = timeit.timeit(time_pca_guided_kmeans, number=num_experiments)

# Measure the time for two_way_kmeans
two_way_kmeans_time = timeit.timeit(time_two_way_kmeans, number=num_experiments)

# Print the results
print(f"Average running time for PCA_guided_kmeans: {pca_guided_kmeans_time / num_experiments:.6f} seconds")
print(f"Average running time for two_way_kmeans: {two_way_kmeans_time / num_experiments:.6f} seconds")
```
